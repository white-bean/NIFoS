### 필지에 . 들어있으면 제거

```python
df['필지'] = df['필지'].apply(lambda x: x.replace(".", ""))
```

### 필지에서 법정동 뽑기

```python
# 필지 주소 안에 '산'이라는 글자와 주번부번이 있는 경우 그 문자열 삭제하기
import re

def splitCode(x):
    pattern = ' *산* *\d+-*\d*$'
    addr = re.sub(pattern, '',str(x))
    return addr
```

```python
df['addr'] = df['필지'].apply(splitCode)
```

```python
# addr 데이터를 공백을 기준으로 분할
addrDF = df['addr'].str.split(" ", expand=True)
addrDF
```

## 산 여부 + 주번/부번 붙이기

```python
# 이제 addr와 법정동명은 쓸 일이 없으니 제거
test.drop(['addr', '법정동명'], axis=1, inplace=True)
test.head()
```

```python
test.shape
```

```python
# (1: 산 아님, 2:산)
import re

def check_mt(x):
    pattern = '산 \d+-*\d*$' # 산 숫자-숫자 인 것만 찾아라
    if re.search(pattern, x):
        return 2
    else:
        return 1
```

```python
# 주번과 부번
def jubu(x):
    pattern = '(\d+)-*(\d*)$' # 숫자(1번이상)-(0번이상)숫자(0번이상) + 끝에서부터
    t = re.search(pattern, x)
    try:
        text = t.group(1).zfill(4) + t.group(2).zfill(4) # 0 채우기
    except:
        print(x)
    return text
```

```python
test['check_mt'] = test['필지'].apply(check_mt)
test['jubu'] = test['필지'].apply(jubu)
test['pnu'] = test['법정동코드'].astype(str) + test['check_mt'].astype(str) + test['jubu'].astype(str)
```

```python
test.head()
```

```python
test.drop(['법정동코드', 'check_mt', 'jubu'], axis=1, inplace=True)
test.head()
```
